##**Diario**
###**TERMINO Y QUE SE?**
####**PROBABILIDAD:** Es una medida de la posibilidad de que ocurra un evento
####**DATO:** Son piezas de información que recogemos del mundo que nos rodea.
####**EVENTO:** Cualquier cosa que pueda suceder en una situación aleatoria.
####**TEORIA DE CONJUNTOS:** Nos ayuda a entender y trabajar con colecciones de objetos, llamados conjuntos.
####**AXIOMAS:** Son como las reglas básicas o fundamentos sobre los que construimos teorías matemáticas.
####**ESTADISTICA:** Se dedica a recoger, analizar y sacar conclusiones a partir de datos.
####**TEORIA DEL CAOS:** Se ocupa de sistemas que parecen desordenados y caóticos, pero que en realidad siguen reglas muy precisas.
####**ESTADISTICA DESCRIPTIVA:** se enfoca en resumir y describir datos para entenderlos mejor.
####**ESTADISTICA INFERENCIAL:** Permite hacer predicciones o generalizaciones sobre una población más grande a partir algunos datos.
####**TEORIA DEL CONJUNTO:** Estudia las propiedades y relaciones que forman un grupo/conjunto
####**POBLACION:** Es un grupo de elementos/personas que cumplen con ciertas caracteristicas especificas
####**MUESTRA:** Es un grupo representativo de la poblacion seleccionado para relizar un analisis de una poblacion mas grande, cumpliendo ciertos requisitos/caracteristicas.
####**MUESTREO:** Es la forma en la que se reliaza la muestra, se encarga de dividir poblaciones, conjuntos, etc.
####**MUESTREO ESTRATIFICADO:** Clasifica datos, personas y objetos en grupos o capas distintos. Se utiliza en combinación con otras herramientas de análisis de datos. Cuando se agrupan datos de diversas fuentes o categorías.

##Comparacion en los metodos de muestreo

```python
methods_comparison = pd.DataFrame({
    "Poblacion General": data["Tiempo de desarrollo (horas)"].mean(),
    "Aleatorio Simple": sample['Tiempo de desarrollo (horas)'].mean(),
    "Aleatorio Sistematico": systematic_sample["Tiempo de desarrollo (horas)"].mean(),
    "Aleatorio Estratificado": stratified_sample["Tiempo de desarrollo (horas)"].mean(),
    "No aleatorio por conveniencia": convenience_sample["Tiempo de desarrollo (horas)"].mean(),
    }, index=["Media de Horas de Desarrollo"])

print("Comparacion de metodos")
display(methods_comparison)
```
